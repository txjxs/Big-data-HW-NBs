```python
import itertools
from collections import Counter
import pandas as pd
```

```python
# Read transactions from the file
with open("/content/browsing.txt", "r") as file:
    transactions = [line.strip().split() for line in file]

```

```python
# Count single item occurrences
item_counts = Counter(item for transaction in transactions for item in transaction)

# Identify frequent single items (support >= 100)
frequent_items = {item for item, count in item_counts.items() if count >= 100}

```

```python
item_counts
```

```python
len(frequent_items)
```

```python
# Generate and count frequent pairs
pair_counts = Counter()
for transaction in transactions:
    filtered_transaction = [item for item in transaction if item in frequent_items]
    pair_counts.update(itertools.combinations(sorted(filtered_transaction), 2))

# Identify frequent pairs (support >= 100)
frequent_pairs = {pair: count for pair, count in pair_counts.items() if count >= 100}

```

```python
pair_counts
```

```python
len(frequent_pairs)
```

```python
# Compute confidence for association rules from pairs
rules = [(X, Y, support / item_counts[X]) for (X, Y), support in frequent_pairs.items()]
rules += [(Y, X, support / item_counts[Y]) for (X, Y), support in frequent_pairs.items()]

# Sort by confidence
rules.sort(key=lambda x: (-x[2], x[0], x[1]))

# Store top 5 rules
df_rules = pd.DataFrame(rules[:5], columns=["X", "Y", "Confidence"])
print("Top 5 Item Pair Rules:")
print(df_rules)

```

```python
# Generate and count frequent triples
triple_counts = Counter()
for transaction in transactions:
    filtered_transaction = [item for item in transaction if item in frequent_items]
    triple_counts.update(itertools.combinations(sorted(filtered_transaction), 3))

# Identify frequent triples (support >= 100)
frequent_triples = {triple: count for triple, count in triple_counts.items() if count >= 100}
```

```python
# Compute confidence for association rules from triples
triple_rules = [((X, Y), Z, support / frequent_pairs.get((X, Y), 1)) for (X, Y, Z), support in frequent_triples.items()]
triple_rules += [((X, Z), Y, support / frequent_pairs.get((X, Z), 1)) for (X, Y, Z), support in frequent_triples.items()]
triple_rules += [((Y, Z), X, support / frequent_pairs.get((Y, Z), 1)) for (X, Y, Z), support in frequent_triples.items()]

# Sort by confidence
triple_rules.sort(key=lambda x: (-x[2], x[0][0], x[0][1], x[1]))

# Store top 5 triple rules
df_triple_rules = pd.DataFrame(triple_rules[:5], columns=["(X, Y)", "Z", "Confidence"])
print("Top 5 Item Triple Rules:")
print(df_triple_rules)
```
